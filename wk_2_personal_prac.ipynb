```python
for num in range(10,20):
    for i in range(2,num):
        if num%i == 0:
            j= num/i
            print ('%r / % r= %r ' % (num,i,j))
            break
    else:
        print (num, '是一个质数')
```

```python
for num in range(10,20):  # 迭代 10 到 20 之间的数字
   for i in range(2,num): # 根据因子迭代
      if num%i == 0:      # 确定第一个因子
         j=num/i          # 计算第二个因子
         print ('%d 等于 %d * %d' % (num,i,j))
         break            # 跳出当前循环
   else:                  # 循环的 else 部分
      print (num, '是一个质数')
```

```python
num = [231,1112,421,642,856]
even=[]
odd=[]
while len(num)>0:
    num_1= num.pop()
    if num_1 %2 == 0:
        even.append(num_1)
    else:
        odd.append(num_1)
        
print("even =",even)
print("odd =",odd)
```
