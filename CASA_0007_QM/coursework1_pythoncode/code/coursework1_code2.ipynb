```python
# calculating VIF
# This function is partially from: https://stackoverflow.com/a/51329496/4667568
from statsmodels.stats.outliers_influence import variance_inflation_factor 
from statsmodels.tools.tools import add_constant

def drop_column_using_vif_(df, thresh=5):
    '''
    Calculates VIF each feature in a pandas dataframe, and repeatedly drop the columns with the highest VIF
    A constant must be added to variance_inflation_factor or the results will be incorrect

    :param df: the pandas dataframe containing only the predictor features, not the response variable
    :param thresh: the max VIF value before the feature is removed from the dataframe
    :return: dataframe with multicollinear features removed
    '''
    while True:
        # adding a constatnt item to the data
        df_with_const = add_constant(df)

        vif_df = pd.Series([variance_inflation_factor(df_with_const.values, i) 
               for i in range(df_with_const.shape[1])], name= "VIF",
              index=df_with_const.columns).to_frame()

        # drop the const
        vif_df = vif_df.drop('const')
        
        # if the largest VIF is above the thresh, remove a variable with the largest VIF
        if vif_df.VIF.max() > thresh:
            # If there are multiple variables with the maximum VIF, choose the first one
            index_to_drop = vif_df.index[vif_df.VIF == vif_df.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            df = df.drop(columns = index_to_drop)
        else:
            # No VIF is above threshold. Exit the loop
            break

    return df
```

```python
# using VIF on the bike_rental dataset

df_predictors_selected_VIF = drop_column_using_vif_(condition_x_final.drop('case_mean_proportion', axis=1))
print("The columns remaining after VIF selection are:")
print(df_predictors_selected_VIF.columns)
```

```python
model_condition_x = sm.OLS(endog=condition_x_final[['case_mean_proportion']], exog=sm.add_constant(df_predictors_selected_VIF)).fit()
model_condition_x.summary()
```

```python
# plot 
plt.scatter(model_condition_x.fittedvalues, model_condition_x.resid)
# adding title and labels
plt.xlabel('Fitted case_mean_proportion')
plt.ylabel('Residual')
plt.title('Residual vs. Fitted Plot of case_mean_proportion')
plt.show()
```

```python
d2 = pd.read_csv('coursework_1_data_2019.csv')
```

```python
d2.info()
```

```python
X = d2['mean_cases']
Y = d2['total_budget']

plt.xlabel('mean_cases')
plt.ylabel('total_budget')

a = plt.plot(X,Y,'.', label='original data')
plt.xlim(0,5500)
plt.ylim(1500,1200000)

slope, intercept, r_value, p_value, std_err = sps.linregress(X, Y) 

b = plt.plot(X, X*slope + intercept, 'r', label = 'fitted line') # Plot the regression line.

plt.legend()
plt.show()

print("y =", round(slope,3), "x +", round(intercept,3))
print("Rsq = ", r_value**2)
print("p-value = ", p_value)
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

simple_regression_model_2 = smf.ols(formula='mean_cases ~ total_budget',data=d2).fit()
print(simple_regression_model_2.summary())
```

```python
# produce regression plots
fig = plt.figure(figsize=(10,8))
fig = sm.graphics.plot_regress_exog(simple_regression_model_2, 'total_budget', fig=fig)
```

```python
plt.figure()
#ols.fit().model is a method to access to the residual.
simple_regression_model_2.resid.plot.density()
plt.show()
```
